```python
# Membaca file dataset
nama_file = "Sales - Marketing customer dataset.csv"
df = pd.read_csv(nama_file)
print("Dataset berhasil dimuat!")
```

```python
# 1c. Menampilkan statistik deskriptif (mean, min, max, kuartil) untuk fitur numerik
print("\n=========================================")
print("        1.C. STATISTIK DESKRIPTIF        ")
print("=========================================")
display(df.describe())
```

```python
# 2a. Menghitung persentase missing value untuk semua kolom
print("=========================================")
print("      2.A. PERSENTASE MISSING VALUE      ")
print("=========================================")

# Rumus: (Jumlah nilai kosong per kolom / Total baris data) * 100
persentase_kosong = (df.isnull().sum() / len(df)) * 100

# Mengubah hasil perhitungan ke bentuk DataFrame agar mudah dibaca & divisualisasikan
df_missing = pd.DataFrame({
    'Nama Kolom': persentase_kosong.index,
    'Persentase Kosong (%)': persentase_kosong.values
})

# Mengurutkan dari persentase yang tertinggi ke terendah
df_missing = df_missing.sort_values(by='Persentase Kosong (%)', ascending=False).reset_index(drop=True)
display(df_missing)
```

```python
# ====================================================================
# TUGAS 3: VISUALISASI DISTRIBUSI VARIABEL TARGET (CHURN)
# ====================================================================

# 1. Menghitung jumlah riil dan persentase untuk setiap kelas target
churn_counts = df['churn'].value_counts()
churn_percentages = df['churn'].value_counts(normalize=True) * 100

print("=========================================")
print("     3.A. DISTRIBUSI KELAS TARGET        ")
print("=========================================")
print(f"Kelas 0 (Pelanggan Tetap) : {churn_counts[0]} data ({churn_percentages[0]:.2f}%)")
print(f"Kelas 1 (Pelanggan Churn) : {churn_counts[1]} data ({churn_percentages[1]:.2f}%)")

# 2. Membuat Visualisasi Diagram Batang
print("\n=========================================")
print("     3.B. VISUALISASI DIAGRAM BATANG     ")
print("=========================================")
plt.figure(figsize=(7, 5))

# Membuat countplot untuk menghitung frekuensi otomatis
sns.countplot(
    x='churn',
    data=df,
    palette='Set2'
)

# Menambahkan judul, label, dan kustomisasi teks sumbu X
plt.title('Visualisasi Keseimbangan Kelas Target (Churn)', fontsize=13, pad=15)
plt.xlabel('Variabel Target (Churn)', fontsize=11)
plt.ylabel('Jumlah Sampel Data', fontsize=11)
plt.xticks([0, 1], ['0 (Tidak Churn)', '1 (Churn)'])

# Menampilkan grafik
plt.tight_layout()
plt.show()
```

```python
# ====================================================================
# TUGAS 4: HEATMAP KORELASI UNTUK FITUR NUMERIK
# ====================================================================

print("=========================================")
print("      4.A. PROSES MATRIKS KORELASI       ")
print("=========================================")

# 1. Memisahkan kolom-kolom yang bertipe numerik saja (int64 dan float64)
fitur_numerik = df.select_dtypes(include=[np.number])

# Jika kolom 'customer_id' ikut masuk, kita drop sementara agar tidak mengacaukan korelasi bisnis
if 'customer_id' in fitur_numerik.columns:
    fitur_numerik = fitur_numerik.drop(columns=['customer_id'])

# 2. Menghitung nilai korelasi Pearson antar fitur numerik
matriks_korelasi = fitur_numerik.corr()

# 3. Membuat Visualisasi Heatmap Korelasi
print("\n=========================================")
print("         4.B. VISUALISASI HEATMAP        ")
print("=========================================")
plt.figure(figsize=(15, 11))

# Membuat heatmap menggunakan seaborn dengan skema warna coolwarm
sns.heatmap(
    matriks_korelasi,
    annot=True,          # Menampilkan angka nilai korelasi di dalam kotak
    fmt='.2f',           # Membatasi 2 angka di belakang koma
    cmap='coolwarm',     # Warna merah (korelasi positif kuat) s.d biru (korelasi negatif kuat)
    linewidths=0.5,      # Memberi garis pembatas tipis antar kotak
    cbar=True            # Menampilkan legenda batang warna di samping
)

plt.title('Heatmap Korelasi Antar Fitur Numerik Terhadap Pola Churn', fontsize=15, pad=20)
plt.tight_layout()
plt.show()
```

```python

# DIRECT MODELING
from sklearn.model_selection import train_test_split

print("=========================================")
print("     1: PENETAPAN VARIABEL X & Y   ")
print("=========================================")

# 1. Sesuai Tugas 1: Menetapkan variabel target (y) yaitu kolom 'churn'
y = df['churn']

# 2. Mengambil seluruh kolom lain sebagai fitur prediktor (X_awal)
X_awal = df.drop(columns=['churn'])

# 3. Filter khusus fitur numerik agar tidak memicu error tipe data pada Scikit-Learn
X_direct = X_awal.select_dtypes(include=['int64', 'float64'])

# 4. Menghapus kolom 'customer_id' jika ada, karena tidak memiliki nilai prediktif bisnis
if 'customer_id' in X_direct.columns:
    X_direct = X_direct.drop(columns=['customer_id'])

# 5. Mengisi missing value secara instan dengan nilai median agar algoritma bisa running
X_direct = X_direct.fillna(X_direct.median())

print(f"-> Kolom target (y) berhasil ditetapkan (Ukuran: {y.shape})")
print(f"-> Kolom prediktor (X) berhasil difilter (Ukuran numerik: {X_direct.shape})")
print(f"Daftar fitur numerik yang digunakan pada Direct Modeling:\n{list(X_direct.columns)}")


print("\n=========================================")
print("       2: TRAIN-TEST SPLIT         ")
print("=========================================")

# Melakukan train-test split untuk membagi data menjadi data latih dan data uji.
# Kita gunakan proporsi standar 80% Data Latih dan 20% Data Uji.
# Parameter stratify=y digunakan agar proporsi kelas churn di data latih dan uji tetap seimbang.
X_train, X_test, y_train, y_test = train_test_split(
    X_direct,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

print("Proses pembagian data berhasil dilakukan!")
print(f"Data Latih - Fitur (X_train)   : {X_train.shape}")
print(f"Data Latih - Target (y_train)  : {y_train.shape}")
print(f"Data Uji   - Fitur (X_test)    : {X_test.shape}")
print(f"Data Uji   - Target (y_test)   : {y_test.shape}")
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

print("=======================================================")
print("  3: INISIALISASI TIGA KATEGORI MODEL          ")
print("=======================================================")

# 1. Kategori Model Konvensional
model_lr = LogisticRegression(max_iter=1000, random_state=42)
print("-> Model Konvensional (Logistic Regression) diinisialisasi.")

# 2. Kategori Model Ensemble Bagging
model_rf = RandomForestClassifier(random_state=42)
print("-> Model Ensemble Bagging (Random Forest) diinisialisasi.")

# 3. Kategori Model Ensemble Voting Classifier (Gabungan LR, KNN, dan SVM)
model_knn = KNeighborsClassifier()
model_svm = SVC(probability=True, random_state=42)

model_voting = VotingClassifier(
    estimators=[('lr', model_lr), ('knn', model_knn), ('svm', model_svm)],
    voting='soft'
)
print("-> Model Ensemble Voting (Voting Classifier) diinisialisasi.\n")


print("=======================================================")
print("  4: PELATIHAN MODEL LANGSUNG (FIT)            ")
print("=======================================================")

# Melatih ketiga model menggunakan data latih yang sudah Anda split sebelumnya
print("1. Sedang melatih Logistic Regression...")
model_lr.fit(X_train, y_train)

print("2. Sedang melatih Random Forest...")
model_rf.fit(X_train, y_train)

print("3. Sedang melatih Voting Classifier (Mohon tunggu sejenak)...")
model_voting.fit(X_train, y_train)

print("[SUKSES] Ketiga model telah selesai dilatih secara langsung!\n")


print("=======================================================")
print("  5: EVALUASI METRIKS & CONFUSION MATRIX       ")
print("=======================================================")

# Membuat fungsi bantuan agar tampilan visualisasi output rapi dan terstruktur
def evaluasi_dan_cetak(model, nama_model):
    # Melakukan prediksi terhadap data uji (X_test) hasil split Anda
    y_pred = model.predict(X_test)

    # Menghitung masing-masing metrik sesuai permintaan soal nomor 5
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    print(f"📊 PERFORMA MODEL BASELINE: {nama_model.upper()}")
    print(f"   - Accuracy  : {acc:.4f} ({acc*100:.2f}%)")
    print(f"   - Precision : {prec:.4f} ({prec*100:.2f}%)")
    print(f"   - Recall    : {rec:.4f} ({rec*100:.2f}%)")
    print(f"   - F1-Score  : {f1:.4f} ({f1*100:.2f}%)")
    print(f"   - Confusion Matrix :\n{cm}")
    print("-" * 55 + "\n")

# Menjalankan fungsi evaluasi untuk menampilkan hasil ketiga model
evaluasi_dan_cetak(model_lr, "Logistic Regression (Konvensional)")
evaluasi_dan_cetak(model_rf, "Random Forest (Ensemble Bagging)")
evaluasi_dan_cetak(model_voting, "Voting Classifier (Ensemble Voting)")
```

```python
# ====================================================================
# MODELING DENGAN PREPROCESSING
# ====================================================================
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

print("=======================================================")
print("  NOMOR 1: IMPUTATION & FEATURE ENGINEERING            ")
print("=======================================================")

# 1. Ambil salinan data awal (df) agar tidak merusak variabel data mentah
df_prep = df.copy()

# 2. Rekayasa Fitur: Mengubah objek string tanggal menjadi durasi hari aktif (tenure_days)
df_prep['signup_date'] = pd.to_datetime(df_prep['signup_date'], errors='coerce')
df_prep['last_purchase_date'] = pd.to_datetime(df_prep['last_purchase_date'], errors='coerce')
df_prep['tenure_days'] = (df_prep['last_purchase_date'] - df_prep['signup_date']).dt.days
df_prep['tenure_days'] = df_prep['tenure_days'].fillna(0).apply(lambda x: x if x >= 0 else 0)

# 3. Drop kolom identitas unik dan teks acak yang tidak berkorelasi dengan target
kolom_drop = ['customer_id', 'signup_date', 'last_purchase_date', 'coupon_code']
df_prep = df_prep.drop(columns=[col for col in kolom_drop if col in df_prep.columns], errors='ignore')

# 4. Imputation Missing Values
# Mengisi fitur numerik yang kosong dengan nilai MEDIAN
fitur_numerik = df_prep.select_dtypes(include=[np.number]).columns.drop('churn', errors='ignore')
for col in fitur_numerik:
    df_prep[col] = df_prep[col].fillna(df_prep[col].median())

# Mengisi fitur kategorikal (teks) yang kosong dengan nilai MODUS
fitur_kategorikal = df_prep.select_dtypes(include=['object']).columns
for col in fitur_kategorikal:
    df_prep[col] = df_prep[col].fillna(df_prep[col].mode()[0])

# 5. One-Hot Encoding untuk merubah kolom teks menjadi representasi biner angka
df_preprocessed = pd.get_dummies(df_prep, columns=fitur_kategorikal, drop_first=True)

print(f"[OK] Preprocessing Selesai. Dimensi dataset baru: {df_preprocessed.shape}\n")


print("=======================================================")
print("  NOMOR 2 & 3: DATA SPLITTING & FEATURE SCALING        ")
print("=======================================================")

# NOMOR 2: Memisahkan prediktor X dan target y, lalu split data (80:20)
X_p = df_preprocessed.drop(columns=['churn'])
y_p = df_preprocessed['churn']

X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(
    X_p, y_p, test_size=0.2, random_state=42, stratify=y_p
)
print(f"-> Ukuran X_train hasil split: {X_train_p.shape}")
print(f"-> Ukuran X_test hasil split : {X_test_p.shape}")

# NOMOR 3: Feature Scaling (Standardization) murni SETELAH proses splitting dilakukan
scaler = StandardScaler()
kolom_kontinu = X_p.select_dtypes(include=['int64', 'float64']).columns

# Fit_transform pada data training, dan transform pada data testing
X_train_p[kolom_kontinu] = scaler.fit_transform(X_train_p[kolom_kontinu])
X_test_p[kolom_kontinu] = scaler.transform(X_test_p[kolom_kontinu])
print("[OK] Standardisasi fitur selesai diaplikasikan.\n")


print("=======================================================")
print("  NOMOR 4: TRAINING MODEL DENGAN DATA PREPROCESSED     ")
print("=======================================================")

# Menginisialisasi ulang tiga kategori model sesuai kriteria penugasan
model_lr_p = LogisticRegression(max_iter=1000, random_state=42)
model_rf_p = RandomForestClassifier(random_state=42)

model_knn_p = KNeighborsClassifier()
model_svm_p = SVC(probability=True, random_state=42)
model_voting_p = VotingClassifier(
    estimators=[('lr', model_lr_p), ('knn', model_knn_p), ('svm', model_svm_p)],
    voting='soft'
)

# Menjalankan proses fitting model ke data training hasil preprocessing
print("-> Sedang memproses latihan Logistic Regression...")
model_lr_p.fit(X_train_p, y_train_p)

print("-> Sedang memproses latihan Random Forest...")
model_rf_p.fit(X_train_p, y_train_p)

print("-> Sedang memproses latihan Voting Classifier (Mohon tunggu)...")
model_voting_p.fit(X_train_p, y_train_p)
print("[OK] Seluruh model berhasil ditraining!\n")


print("=======================================================")
print("  NOMOR 5: EVALUASI METRIKS SKENARIO PREPROCESSING    ")
print("=======================================================")

def hitung_metrik_prep(model, nama_model):
    # Prediksi data uji menggunakan data yang sudah discaling
    y_pred_p = model.predict(X_test_p)

    # Hitung metrik evaluasi
    acc = accuracy_score(y_test_p, y_pred_p)
    prec = precision_score(y_test_p, y_pred_p, zero_division=0)
    rec = recall_score(y_test_p, y_pred_p)
    f1 = f1_score(y_test_p, y_pred_p)
    cm = confusion_matrix(y_test_p, y_pred_p)

    print(f"📊 PERFORMA MODEL PREPROCESSED: {nama_model.upper()}")
    print(f"   - Accuracy  : {acc:.4f} ({acc*100:.2f}%)")
    print(f"   - Precision : {prec:.4f} ({prec*100:.2f}%)")
    print(f"   - Recall    : {rec:.4f} ({rec*100:.2f}%)")
    print(f"   - F1-Score  : {f1:.4f} ({f1*100:.2f}%)")
    print(f"   - Confusion Matrix :\n{cm}")
    print("-" * 55 + "\n")

# Menampilkan output evaluasi performa nomor 5
hitung_metrik_prep(model_lr_p, "Logistic Regression (Prep)")
hitung_metrik_prep(model_rf_p, "Random Forest (Prep)")
hitung_metrik_prep(model_voting_p, "Voting Classifier (Prep)")
```

### 4. Hyperparameter Tuning
Pada tahap ini, kita melakukan proses optimasi parameter (*hyperparameter tuning*) pada model **Random Forest Classifier** menggunakan metode `GridSearchCV` dengan *5-Fold Cross Validation* untuk mencari kombinasi parameter yang menghasilkan performa paling optimal:

1. **Nomor 1 (Penentuan Parameter):** Menentukan daftar parameter grid yang akan diuji, yaitu `n_estimators` (jumlah pohon), `max_depth` (kedalaman maksimum pohon), dan `min_samples_split` (jumlah minimum sampel untuk split node).
2. **Nomor 2 (Penerapan GridSearchCV):** Mengonfigurasi `GridSearchCV` dengan melampirkan model dasar, parameter grid, serta `cv=5` (5-Fold Cross Validation). Proses evaluasi dioptimalkan menggunakan *scoring* F1-Score karena sebaran kelas target *imbalanced*.
3. **Nomor 3 (Model Training / Fit):** Melatih (*fitting*) objek `GridSearchCV` pada data training hasil preprocessing (`X_train_p` dan `y_train_p`) untuk mencari kombinasi terbaik secara otomatis.
4. **Nomor 4 (Mengekstrak Model Terbaik):** Mengambil dan mengisolasi arsitektur model dengan parameter terbaik yang berhasil ditemukan oleh sistem (`best_estimator_`).
5. **Nomor 5 (Model Pengujian / Predict):** Menggunakan model terbaik hasil tuning tersebut untuk melakukan prediksi label target pada data testing (`X_test_p`).
6. **Nomor 6 (Evaluasi Performa Final):** Menghitung dan menampilkan metrik performa akhir yang meliputi *Accuracy, Precision, Recall, F1-Score*, serta menyajikan struktur *Confusion Matrix* setelah proses tuning selesai dilakukan.

```python
# ====================================================================
# HYPERPARAMETER TUNING
# ====================================================================
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

print("=======================================================")
print("  NOMOR 1 & 2: SETUP PARAMETER GRID & GRIDSEARCHCV     ")
print("=======================================================")

# NOMOR 1: Menentukan daftar parameter grid yang akan dieksplorasi
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10]
}
print(f"-> Parameter grid yang ditentukan:\n   {param_grid}\n")

# Inisialisasi base model Random Forest
rf_base = RandomForestClassifier(random_state=42)

# NOMOR 2: Menerapkan fungsi GridSearchCV dengan Cross-Validation = 5
grid_search = GridSearchCV(
    estimator=rf_base,
    param_grid=param_grid,
    cv=5,
    scoring='f1', # Mengoptimalkan F1-Score karena datanya imbalanced
    n_jobs=-1,    # Menggunakan core maksimal agar proses pencarian lebih cepat
    verbose=1
)


print("=======================================================")
print("  NOMOR 3 & 4: TRAINING (FIT) & EKSTRAKSI MODEL TERBAIK")
print("=======================================================")

# NOMOR 3: Melatih GridSearchCV pada data latih hasil preprocessing
print("-> Sedang memproses pencarian kombinasi parameter optimal (Mohon tunggu)...")
grid_search.fit(X_train_p, y_train_p)

# NOMOR 4: Mengekstrak model terbaik hasil temuan ke dalam variabel baru
best_rf_model = grid_search.best_estimator_
print(f"\n[OK] Parameter terbaik yang berhasil ditemukan:\n     {grid_search.best_params_}\n")


print("=======================================================")
print("  NOMOR 5 & 6: PREDIKSI DATA UJI & EVALUASI PERFORMA   ")
print("=======================================================")

# NOMOR 5: Melakukan prediksi terhadap data uji (X_test_p) menggunakan model terbaik
print("-> Melakukan prediksi final pada data testing...")
y_pred_tuned = best_rf_model.predict(X_test_p)

# NOMOR 6: Menghitung seluruh matriks evaluasi sesuai instruksi soal
acc_tuned = accuracy_score(y_test_p, y_pred_tuned)
prec_tuned = precision_score(y_test_p, y_pred_tuned, zero_division=0)
rec_tuned = recall_score(y_test_p, y_pred_tuned)
f1_tuned = f1_score(y_test_p, y_pred_tuned)
cm_tuned = confusion_matrix(y_test_p, y_pred_tuned)

print(f"\n📊 HASIL EVALUASI FINAL MODEL: RANDOM FOREST (TUNED)")
print(f"   - Accuracy  : {acc_tuned:.4f} ({acc_tuned*100:.2f}%)")
print(f"   - Precision : {prec_tuned:.4f} ({prec_tuned*100:.2f}%)")
print(f"   - Recall    : {rec_tuned:.4f} ({rec_tuned*100:.2f}%)")
print(f"   - F1-Score  : {f1_tuned:.4f} ({f1_tuned*100:.2f}%)")
print(f"   - Confusion Matrix :\n{cm_tuned}")
print("-" * 55)
```
